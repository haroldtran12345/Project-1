```python
import yfinance as yf
import pandas as pd
output_data_file = "output_data/carstockprices.csv"
```

```python
TSLA = yf.Ticker("TSLA")
TM = yf.Ticker("TM")
F = yf.Ticker("F")
GM = yf.Ticker("GM")
HMC = yf.Ticker("HMC")
DDAIF = yf.Ticker("DDAIF")
VWAGY = yf.Ticker("VWAGY")
FUJHY = yf.Ticker("FUJHY")
MZDAY = yf.Ticker("MZDAY")
DODGX = yf.Ticker("DODGX")
```

```python
tsla_historical = TSLA.history(start="2014-01-01", end="2021-07-01", interval="1d")
tm_historical = TM.history(start="2014-01-01", end="2021-07-01", interval="1d")
f_historical = F.history(start="2014-01-01", end="2021-07-01", interval="1d")
gm_historical = GM.history(start="2014-01-01", end="2021-07-01", interval="1d")
hmc_historical = HMC.history(start="2014-01-01", end="2021-07-01", interval="1d")
ddaif_historical = DDAIF.history(start="2014-01-01", end="2021-07-01", interval="1d")
vwagy_historical = VWAGY.history(start="2014-01-01", end="2021-07-01", interval="1d")
fujhy_historical = FUJHY.history(start="2014-01-01", end="2021-07-01", interval="1d")
mzday_historical = MZDAY.history(start="2014-01-01", end="2021-07-01", interval="1d")
dodgx_historical = DODGX.history(start="2014-01-01", end="2021-07-01", interval="1d")
```

```python
tsla_historical
```

```python
f_historical["Ticker"]="F"
tm_historical["Ticker"]="TM" 
gm_historical["Ticker"]="GM"
hmc_historical["Ticker"]="HMC"
ddaif_historical["Ticker"]="DDAIF"
vwagy_historical["Ticker"]="VWAGY"
tsla_historical["Ticker"]="TSLA"
fujhy_historical["Ticker"]="FUJHY"
mzday_historical["Ticker"]="MZDAY"
dodgx_historical["Ticker"]="DODGX"
```

```python
merged_df = pd.concat([f_historical, tm_historical, gm_historical, hmc_historical, ddaif_historical, vwagy_historical,tsla_historical, fujhy_historical, mzday_historical, dodgx_historical ])
merged_df
```

```python
del merged_df['Dividends']
```

```python
del merged_df['Stock Splits']
```

```python
merged_df.reset_index()
```

```python
merged_df.to_csv(output_data_file, encoding= "utf-8", index = True, header= True)
```
